```python
Ams:4
# Import the required libraries
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
my_db = client['my_database']
my_collection = my_db['my_collection']

# Insert one record into the collection
document1 = {"name": "Alice", "age": 25, "city": "Los Angeles"}
result1 = my_collection.insert_one(document1)
print("One document inserted with ID:", result1.inserted_id)

# Insert many records into the collection
documents2 = [
    {"name": "Bob", "age": 35, "city": "San Francisco"},
    {"name": "Charlie", "age": 28, "city": "Seattle"},
    {"name": "David", "age": 31, "city": "Chicago"}
]
result2 = my_collection.insert_many(documents2)
print("Many documents inserted with IDs:", result2.inserted_ids)

# Find and print inserted records using find()
print("Inserted records using find():")
for document in my_collection.find():
    print(document)

# Find and print inserted records using find_one()
print("Inserted records using find_one():")
document3 = my_collection.find_one({"name": "Alice"})
print(document3)

# Close the MongoDB connection
client.close()
```

```python
Ans:5
# Import the required libraries
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
my_db = client['my_database']
my_collection = my_db['my_collection']

# Find documents that match a specific filter
filter = {"age": {"$gte": 30}}  # Retrieve documents where 'age' is greater than or equal to 30
cursor = my_collection.find(filter)

# Iterate over the cursor and print the retrieved documents
print("Documents retrieved using find():")
for document in cursor:
    print(document)

# Close the MongoDB connection
client.close()
```

```python
Ans:6
# Import the required libraries
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
my_db = client['my_database']
my_collection = my_db['my_collection']

# Find documents and sort them by age in ascending order
cursor1 = my_collection.find().sort("age", 1)

# Iterate over the cursor and print the sorted documents
print("Documents sorted by age in ascending order:")
for document in cursor1:
    print(document)

# Find documents and sort them by age in descending order
cursor2 = my_collection.find().sort("age", -1)

# Iterate over the cursor and print the sorted documents
print("Documents sorted by age in descending order:")
for document in cursor2:
    print(document)

# Close the MongoDB connection
client.close()
```
